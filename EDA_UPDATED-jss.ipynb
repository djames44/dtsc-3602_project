```python
# REPLACE your "violent_crimes" filter (which actually filters to sexual-only) with a binary flag.
# 1 = violent, 0 = non-violent (rest of crimes).
violent_crimes = [
    "RAPE, FORCIBLE",
    "RAPE, ATTEMPTED",
    "SEXUAL PENETRATION W/ FOREIGN OBJECT",
    "UNLAWFUL SEX (INCLUDING MUTUAL CONSENT)",
    "ORAL COPULATION",
    "SODOMY/SEXUAL CONTACT B/W PENIS/ANUS",
    "BATTERY WITH SEXUAL CONTACT",
    "HUMAN TRAFFICKING - COMMERCIAL SEX ACTS",
    "HUMAN TRAFFICKING - INVOLUNTARY SERVITUDE",
    "PIMPING"
]

#crime_df = crime_df[crime_df['Crm Cd Desc'].isin(violent_crimes)]
# standardize crime description before flagging
crime_df["Crm Cd Desc"] = crime_df["Crm Cd Desc"].astype(str).str.upper().str.strip()
crime_df["SexualAssault"] = crime_df["Crm Cd Desc"].isin(violent_crimes).astype(int)
print("SexualAssault counts (1/0):")
print(crime_df["SexualAssault"].value_counts())

```

```python
# select columns of interest 
crime_df = crime_df[['DR_NO', 'DATE OCC', 'TIME OCC', 'AREA', 'AREA NAME',
       'Rpt Dist No', 'Crm Cd', 'Crm Cd Desc', 'Mocodes',
       'Vict Age', 'Vict Sex', 'Vict Descent', 'Premis Cd', 'Premis Desc',
       'Weapon Used Cd', 'Weapon Desc', 'Status', 'Status Desc', 'LAT',
       'LON', 'SexualAssault']]
crime_df.shape
```

```python
# drop if missing or null value, sort by condition
#crime_df = crime_df.dropna(subset=["Crm Cd Desc", "Vict Age", "Vict Sex"])
#crime_df = crime_df[(crime_df["Vict Age"] >= 10) & (crime_df["Vict Age"] <= 100)]
#crime_df.info()

#new
## Drop rows only if crime type is missing (can't classify crime otherwise)

before = len(crime_df)
crime_df = crime_df.dropna(subset=["Crm Cd Desc"])
print("Dropped rows missing Crm Cd Desc:", before - len(crime_df))

# fill categoricals with 'Unknown' (don't throw away rows)
for c in ["Vict Sex", "Vict Descent", "Premis Desc", "Weapon Desc", "Status Desc", "AREA NAME"]:
    miss = crime_df[c].isna().sum()
    if miss:
        print(f"{c}: filling {miss} missing with 'Unknown'")
    crime_df[c] = crime_df[c].fillna("Unknown").astype(str).str.strip()

# normalize Vict Sex - keeps it consistant
crime_df["Vict Sex"] = (crime_df["Vict Sex"].str.upper()
                        .replace({"X": "UNKNOWN", "U": "UNKNOWN", "NAN": "UNKNOWN"}))

# validate Vict Age: allow NaN, drop impossible ages
crime_df["Vict Age"] = pd.to_numeric(crime_df["Vict Age"], errors="coerce")
invalid_age = (crime_df["Vict Age"] < 10) | (crime_df["Vict Age"] > 100)
print("Invalid ages removed:", invalid_age.sum())
crime_df = crime_df.loc[~invalid_age].copy()

crime_df.info()
```

```python
# ========= 6) (ADD) Location cleaning â€” needed for neighborhood/place effects =========

# make sure LAT/LON are numbers 
# (needed for mapping and neighborhood analysis)
for col in ["LAT", "LON"]:
    crime_df[col] = pd.to_numeric(crime_df[col], errors="coerce")
# flag invalid rows where coords are (0,0) 
# (not real LA locations, just placeholders)
bad_xy = (crime_df["LAT"] == 0) | (crime_df["LON"] == 0)
# set those invalid coords to NaN 
# (so they can be dropped in the next step)
crime_df.loc[bad_xy, ["LAT", "LON"]] = pd.NA


# drop rows with missing/invalid coords
# (important because we need real locations 
# to measure neighborhood concentration)
b4 = len(crime_df)
crime_df = crime_df.dropna(subset=["LAT", "LON"])

# show how many rows were removed
print("Rows dropped for invalid/missing coords:", b4 - len(crime_df))
```

```python
#NEW ADDED: --> To measure concentration of prior incidents
#With datetime_occurrence, you can say:This incident happened on June 3rd at 11:45 PM in Hollywood, and in the 90 days before that, there were 5 other sexual assaults in the same reporting district. --> needed for the prior incidents for our model and reserch question!!
crime_df["DATE_OCC_dt"] = pd.to_datetime(crime_df["DATE OCC"], errors="coerce")
t = pd.to_numeric(crime_df["TIME OCC"], errors="coerce").fillna(0).astype(int)
hour = (t // 100).clip(0, 23)
minute = (t % 100).clip(0, 59)
crime_df["datetime_occurrence"] = pd.to_datetime(
    crime_df["DATE_OCC_dt"].dt.date.astype(str) + " " + hour.astype(str) + ":" + minute.astype(str),
    errors="coerce"
)


print("Missing datetime_occurrence:", crime_df["datetime_occurrence"].isna().sum())
```

# Exploratory Data Analysis (EDA)
**********

The EDA uses the **cleaned `crime_df`** created in the first part of the notebook and focuses on:
- Victim **gender** distributions
- Counts by **neighborhood**
- **Time trends** (monthly/yearly)
- Cross-tabs: **Gender Ã— Crime Type**, **Gender Ã— Neighborhood**

## 1) Data Loading / Focus Filter

```python
# Data Loading
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas.api.types import is_datetime64_any_dtype

# Focus filter:
# - If violent_crimes list exists, keep only those crimes
# - Otherwise if SexualAssault column exists, keep flagged rows
# - Otherwise keep everything
if 'violent_crimes' in globals():
    df_eda = crime_df[crime_df['Crm Cd Desc'].isin(violent_crimes)].copy()
    print(f"Subset applied: {len(df_eda):,} rows in violent_crimes list")
elif 'SexualAssault' in crime_df.columns:
    df_eda = crime_df[crime_df['SexualAssault'] == 1].copy()
    print(f"Subset applied: {len(df_eda):,} rows where SexualAssault == 1")
else:
    df_eda = crime_df.copy()
    print("No focus filter found (violent_crimes/SexualAssault). Using all rows.")

# Show dimensions and first 5 rows
print(f"Rows: {len(df_eda):,} | Columns: {df_eda.shape[1]}")
display(df_eda.head(5))
```

## 2) Victim Gender Distribution

```python
# Victim Gender Distribution
gender_col_candidates = ['Vict Sex','Vict_Sex','victim_sex','victim_gender','Vict Gender']
gender_col = next((c for c in gender_col_candidates if c in df_eda.columns), None)
if gender_col is None:
    print('No victim gender column found among:', gender_col_candidates)
else:
    # Count genders and calculate percentages
    counts = df_eda[gender_col].value_counts(dropna=False)
    pct = (counts / counts.sum() * 100).round(2)

    # Print counts and percentages
    print('Counts by gender:')
    display(counts)
    print('\nPercent by gender:')
    display(pct)

    # Bar chart of gender distribution
    plt.figure()
    counts.plot(kind='bar')
    plt.title('Victim Gender Distribution (Focus Crimes)')
    plt.ylabel('Count')
    plt.xlabel(gender_col)
    plt.xticks(rotation=0)
    plt.tight_layout()
    plt.show()
```

## 3) Focus Crime Types â€” Distribution

```python
# Crime Type Distribution
crime_col = 'Crm Cd Desc' if 'Crm Cd Desc' in df_eda.columns else None
if crime_col is None:
    print('Crime description column not found (expected "Crm Cd Desc").')
else:
    # Count crime types
    crime_counts = df_eda[crime_col].value_counts()
    display(crime_counts)

    # Bar chart of crime types
    plt.figure(figsize=(14,7))
    ax = crime_counts.plot(kind='bar')

    plt.title('Distribution of Focus Crime Types')
    plt.ylabel('Count')
    plt.xlabel('Crime Type')

    # Rotate labels to fit better
    plt.xticks(rotation=45, ha='right')

    # Y-axis increments of 1000
    max_val = crime_counts.max()
    plt.yticks(np.arange(0, max_val + 1000, 1000))

    plt.tight_layout()
    plt.show()
```

## 4) Neighborhood Counts (Focus Crimes)

```python
# Neighborhood Counts
neigh_candidates = ['AREA_NAME','Neighborhood','neighborhood','AREA NAME','Reporting District']
neigh_col = next((c for c in neigh_candidates if c in df_eda.columns), None)
if neigh_col is None:
    print('Neighborhood column not found among:', neigh_candidates)
else:
    # Count top 25 neighborhoods
    nbr_counts = df_eda[neigh_col].value_counts().head(25)
    display(nbr_counts)

    # Bar chart of neighborhoods
    plt.figure(figsize=(10,5))
    nbr_counts.plot(kind='bar')
    plt.title('Top Neighborhoods by Focus Crime Count (Top 25)')
    plt.ylabel('Count')
    plt.xlabel(neigh_col)
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```

## 5) Time Trends (Monthly / Yearly)

```python
# Time Trends
date_candidates = ['DATE OCC','DATE_OCC','Date','date','Rpt Date','DATE RPTD','Reported Date']
date_col = next((c for c in date_candidates if c in df_eda.columns), None)
if date_col is None:
    print('No date column found among:', date_candidates)
else:
    # Convert to datetime if not already
    if not is_datetime64_any_dtype(df_eda[date_col]):
        try:
            df_eda[date_col] = pd.to_datetime(df_eda[date_col], errors='coerce')
        except Exception as e:
            print('Failed to parse dates:', e)

    # Drop rows with NaT dates just in case
    tmp = df_eda.dropna(subset=[date_col]).copy()
    if not tmp.empty:
        # Set date as index and resample monthly and yearly
        tmp = tmp.set_index(date_col).sort_index()
        monthly = tmp.resample('M').size()
        yearly = tmp.resample('Y').size()

        # Plot monthly trend
        plt.figure()
        monthly.plot()
        plt.title('Monthly Focus Crime Counts')
        plt.ylabel('Count')
        plt.xlabel('Month')
        plt.tight_layout()
        plt.show()

        # Plot yearly trend
        plt.figure()
        yearly.plot()
        plt.title('Yearly Focus Crime Counts')
        plt.ylabel('Count')
        plt.xlabel('Year')
        plt.tight_layout()
        plt.show()
    else:
        print('All dates are NaT after parsing; skipping time trends.')
```

## 6) Gender x Crime Type

```python
# Gender x Crime Type Bar Graph
if gender_col is not None and crime_col is not None:
    ct = pd.crosstab(df_eda[gender_col], df_eda[crime_col])
    display(ct)

    # Stacked bar plot
    plt.figure(figsize=(14,6))
    ax = (ct.T).plot(kind='bar', stacked=True, figsize=(14,6))

    plt.title('Gender Ã— Crime Type (Counts)')
    plt.ylabel('Count')
    plt.xlabel('Crime Type')
    plt.xticks(rotation=45, ha='right')

    # Legend outside chart
    plt.legend(title=gender_col, bbox_to_anchor=(1.05, 1), loc='upper left')

    # Y-axis increments of 1000
    max_val = (ct.T).sum(axis=1).max()
    plt.yticks(np.arange(0, max_val + 1000, 1000))

    plt.tight_layout()
    plt.show()
else:
    print('Skipping Gender Ã— Crime Type plot (missing gender or crime column).')
```

## 7) Victim Gender x Crime Type (Heatmap)

```python
# Victim Gender Ã— Crime Type Heatmap
import seaborn as sns

if gender_col is not None and crime_col is not None:
    ct_heat = pd.crosstab(df_eda[gender_col], df_eda[crime_col], normalize="index")
    
    plt.figure(figsize=(12, max(4, 0.6 * len(ct_heat.index))))
    
    sns.heatmap(ct_heat, annot=True, fmt=".2f", cmap="coolwarm")
    plt.title("Heatmap: Victim Gender Ã— Crime Type (Row-normalized)")
    plt.ylabel("Victim Gender")
    plt.xlabel("Crime Type")
    
    plt.xticks(rotation=45, ha="right")
    plt.yticks(rotation=0)  
    plt.tight_layout()
    plt.show()
else:
    print("Skipping Gender Ã— Crime Type heatmap (missing column).")
```

## 8) Crime Type x Neighborhood (Heatmap)

```python
# Crime Type Ã— Neighborhood Heatmap
if gender_col is not None and crime_col is not None:
    ct_heat = pd.crosstab(df_eda[gender_col], df_eda[crime_col], normalize="index")

    # Heatmap with dynamic height
    plt.figure(figsize=(12, max(4, 0.6 * len(ct_heat.index))))
    sns.heatmap(ct_heat, annot=True, fmt=".2f", cmap="coolwarm")

    plt.title("Heatmap: Victim Gender Ã— Crime Type (Row-normalized)")
    plt.ylabel("Victim Gender")
    plt.xlabel("Crime Type")

    plt.xticks(rotation=45, ha='right')
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.show()
else:
    print("Skipping Gender Ã— Crime Type heatmap (missing column).")
```

## 9)Crime type (sexual assult focus x Neighborhood Heatmap)

## 10) Female, Male, and Unkown for each race who experinced sexual assults.


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Use all sex categories (F, M, UNKNOWN)
df_race_sex = df_eda_sa.copy()

# Get top 8 race categories for readability
top_races = df_race_sex["Vict Descent"].value_counts().head(8).index
df_race_sex = df_race_sex[df_race_sex["Vict Descent"].isin(top_races)]

# Count per race Ã— sex
race_sex_counts = (
    df_race_sex.groupby(["Vict Descent","Vict Sex"])
    .size()
    .reset_index(name="Count")
)

plt.figure(figsize=(10,6))
sns.barplot(
    data=race_sex_counts,
    x="Vict Descent",
    y="Count",
    hue="Vict Sex",
    palette="Set2"
)

plt.title("Victim Race/Ethnicity by Sex (Top 8 Categories, incl. Unknown)")
plt.xlabel("Race/Ethnicity (Vict Descent)")
plt.ylabel("Number of Victims")
plt.xticks(rotation=45, ha="right")
plt.legend(title="Victim Sex")
plt.tight_layout()
plt.show()
```

## Limitations and Bias

- In cases of sexual assault, many victims do not report crimes, including sexual assault incidents. This applies to both women and men of all ages. Underreporting leads to missing cases, making the dataset less accurate than desired.  

- Another limitation is the presence of missing demographics within the data, such as "Unknown" values in victim sex and other reported characteristics.  

- A bias in the dataset comes from its starting point in 2020. There may have been more or fewer cases before this time, so the dataset does not reflect long-term trends.  

- Geographic bias is another limitation. Some records had missing values in the "LAT" and "LON" columns, which creates gaps in location-based analysis.  

- Another bias is systemic bias. Historical and structural inequalitiesâ€”such as race, immigration status, and economic conditionsâ€”can influence who reports crimes and how cases are handled. These broader systemic factors are not captured in the dataset.
