```python
import pandas as pd
import geopandas as gpd
crime_path = "/Users/dhanyaelsajames/Downloads/Crime_Data_from_2020_to_Present.csv"
unemp_path = "/Users/dhanyaelsajames/Downloads/Unemployment_-5936650746174720833.geojson"

crime = pd.read_csv(crime_path).dropna(subset=["LAT", "LON"])
crime_gdf = gpd.GeoDataFrame(
    crime,
    geometry=gpd.points_from_xy(crime["LON"], crime["LAT"]),
    crs="EPSG:4326"  # WGS84 lat/lon
)

unemp = gpd.read_file(unemp_path)  # should include columns: OBJECTID, tract, name, unemployment, sup_dist, csa, spa

# 3) Align CRS
if unemp.crs != crime_gdf.crs:
    unemp = unemp.to_crs(crime_gdf.crs)

joined = gpd.sjoin(crime_gdf, unemp, how="left", predicate="within")

# 5) Quick sanity check
print(joined[["DR_NO", "AREA NAME", "unemployment"]].head())



```

```python
joined.drop(columns="geometry").to_csv("/Users/dhanyaelsajames/Downloads/Crime_with_Unemployment.csv", index=False)
print("✅ Spatial join complete — file saved as Crime_with_Unemployment.csv")
```

```python
# Check how many crimes were matched with unemployment data
total = len(joined)
matched = joined['unemployment'].notna().sum()
print(f"Matched {matched} of {total} crimes ({matched/total:.2%}) with unemployment data")
```
